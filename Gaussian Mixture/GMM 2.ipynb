```python
from sklearn import datasets,mixture,cluster
from sklearn.metrics import silhouette_score,adjusted_rand_score
import matplotlib.pyplot as plt
```

```python
x,y=datasets.make_blobs(n_samples=2000,centers=3,cluster_std=0.5,random_state=3)#to make clusters dataset
```

```python
plt.scatter(x[:,0],x[:,1],c=y,edgecolors='black')
```

```python
gmm=mixture.GaussianMixture(n_components=3)
pred=gmm.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=pred,edgecolors='black')
```

# Using internal cluster validation index

```python
silhouette_score(x,pred)
```

```python
adjusted_rand_score(pred,y)
```

```python
k_clusters=[2,3,4,5,6,7,8]
scores=[]
```

```python
for k in k_clusters:
    gmm=mixture.GaussianMixture(n_components=k)
    pred=gmm.fit_predict(x)
    scores.append(silhouette_score(x,pred))
    
plt.plot(k_clusters,scores, color='b', marker='o', linestyle='dashed',linewidth=1, markersize=5)
plt.xlabel="k_clusters"
plt.ylabel="score"
plt.title="Silhouette score"
plt.show()
```
